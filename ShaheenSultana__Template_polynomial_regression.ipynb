# Polynomial Regression

y=b0+b1x1+bax1^2+b3x1^3-----bnx1^n
where n is the degree of polynomial
polynomial linear regression used for curve linear values when y is dependent on x
note underfitting :if degree of polynomial is less then curve may underfit the data leaving more datapoints outside the curve
note overfitting:if degree of polynomial is very high then the curve may overfit the data and might be too much fitting training data making it unfit for the unseen data



## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```
dataset=pd.read_csv('Corn_Yield.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print(x)
print("")
print(y)
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg =LinearRegression()
lin_reg.fit(x,y)# we are not predicting results so we are not splittiong dataset into train & test
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=2)# our data of degree 2 it is giving fittting  if we increase degree it is underfits the data
x_poly=pf.fit_transform(x)
poly_reg = LinearRegression()
poly_reg.fit(x_poly,y)
```

```
print(x_poly)
```

## Visualising the Linear Regression results

```
plt.scatter(x,y,color='blue')
plt.plot(x,lin_reg.predict(x),color ='red')
plt.title(' Linear-Regression for Crop Yield')
plt.xlabel('fertilizer used')
plt.ylabel('Corn Yield')
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(x,y,color='blue')
plt.plot(x,poly_reg.predict(x_poly),color ='red')
plt.title('Polynomial Linear-Regression for Crop Yield')
plt.xlabel('fertilizer used')
plt.ylabel('Corn Yield')
plt.show()
```

## Predicting a new result with Linear Regression

```
lin_reg.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
poly_reg.predict(pf.fit_transform([[6.5]]))
```
