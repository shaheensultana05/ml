# Multiple Linear Regression

## Importing the libraries

```
import numpy as np  #to work with arrays
import pandas as pd #to work with datasets
import matplotlib.pyplot as plt #to visualize the data
```

## Importing the dataset

```
dataset=pd.read_csv("Successful_Startups.csv")
x=dataset.iloc[:,:-1].values#uploading all columns except last column independent variables
y=dataset.iloc[:,-1].values#uploading only last column depandent variable
print(x)
print("")
print(y)
```

## Encoding categorical data

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
x=np.array(ct.fit_transform(x))
print(x)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x ,y ,test_size=0.2, random_state=0)
print(x_train)
print("")
print(x_test)
print("")
print(y_train)
print("")
print(y_test)
```

## Training the Multiple Linear Regression model on the Training set

```
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train) # to fit hyper plane data in plane
```

## Predicting the Test set results

```
y_predict = regr.predict(x_test)# predicting x_test result  #print(y_test) print(y_predict)
np.set_printoptions(precision=2)# want only 2 decimal points
print(np.concatenate((y_predict.reshape(len(y_predict),1),y_test.reshape(len(y_test),1)),1)) # in output actual output(statsups file data) and predicted output is there
```

## Visualize the result to check accuracy

```
# we are seeing how much y_predict change on  y_test change
plt.figure(figsize =(8,6))
plt.scatter(y_test,y_predict,color ='blue',alpha =0.7) #on x-axis x_test & on y_axis y_predict  alphea is transparency
plt.xlabel("Actual values") # show in output as labels
plt.ylabel(" Predicted values")
plt.title("Actual values vs Predicted Values")
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color ='red',alpha=0.7 ,linestyle='--') #linestyle is used to style a line - for ____________________ -- for _ _ _ _ _ _ _ _ line
plt.show() #showing complete plot result
```
